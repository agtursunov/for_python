```python
# Task 1
def first_plot():
    x = np.linspace(-10, 10, 100)
    y = x**2 - 4*x + 4

    plt.plot(x, y)
    
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('the graph: f(x) = x^2 - 4x + 4')
    plt.grid(True)
    
    plt.show()
first_plot()
```

```python
# Task 2
def cos_sin_plots():
    x = np.arange(0, 2 * np.pi, np.pi/10)
    y_sin = np.sin(x)
    y_cos = np.cos(x)

    plt.plot(x, y_sin, label="sin(x)", linestyle='-', color='b', marker='o')
    plt.plot(x, y_cos, label="cos(x)", linestyle='--', color='r', marker='x')
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Sin and Cos functions:')
    plt.legend()
    plt.grid(True)
    plt.show()
cos_sin_plots()
```

```python
# Task 3
def x_pow_3(axes, x):
    y1 = x**3
    
    axes[0, 0].plot(x, y1, color='g')
    
    axes[0, 0].set_title('f(x) = x^3')
    axes[0, 0].set_xlabel('x')
    axes[0, 0].set_ylabel('f(x)')
    axes[0, 0].grid()
    

def sin_plot(axes, x):
    y2 = np.sin(x)
    
    axes[0, 1].plot(x, y2, color='b')
    
    axes[0, 1].set_title('f(x) = sin(x)')
    axes[0, 1].set_xlabel('x')
    axes[0, 1].set_ylabel('f(x)')
    axes[0, 1].grid()


def exp_plot(axes, x):
    y3 = np.exp(x)
    
    axes[1, 0].plot(x, y3, color='r')
    
    axes[1, 0].set_title('f(x) = exp(x)')
    axes[1, 0].set_xlabel('x')
    axes[1, 0].set_ylabel('f(x)')
    axes[1, 0].grid()

def log_graph(axes, x):
    x_positives = np.array([element for element in x if element + 1 > 0])
    y4 = np.log(x_positives + 1)
    
    axes[1, 1].plot(x_positives, y4, color='m')
    
    axes[1, 1].set_title('f(x) = log(x+1)')
    axes[1, 1].set_xlabel('x')
    axes[1, 1].set_ylabel('f(x)')
    axes[1, 1].grid()

def main_func():
    x = np.linspace(-10, 10, 100)
    
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))
    
    fig.suptitle('The four plots:')
    x_pow_3(axes, x)
    sin_plot(axes, x)
    exp_plot(axes, x)
    log_graph(axes, x)
    plt.tight_layout()
    plt.show()

main_func()

```

```python
# Task 4
def scater_random():
    x = np.random.uniform(0, 10, 100)
    y = np.random.uniform(0, 10, 100)

    plt.scatter(x, y, color='r', marker='*')

    plt.title('Scatter Plot with random values')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)

    plt.show()
    
scater_random()
```

```python
# Task 5
def histogram_normal():
    data = np.random.normal(0, 1, 1000)

    plt.hist(data, bins=30, alpha=0.7759, color='k')
    
    plt.title('Histogram with normal random data')
    plt.xlabel('Value')
    plt.ylabel('')

    plt.show()
histogram_normal()
```

```python
# Task 6
def graph_3d():
    x = np.linspace(-5, 5, 10)
    y = np.linspace(-5, 5, 10)
    xx, yy = np.meshgrid(x, y)
    Z = np.cos(xx**2 + yy**2)

    plt.figure(figsize=(12, 8))
    ax = plt.axes(projection='3d')

    ax.plot_surface(xx, yy, Z, cmap='viridis')

    ax.set_title('The 3D Graph')

    plt.show()
graph_3d()

```

```python
# Task 7
def bar_chart():
    products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
    sales = [200, 150, 250, 175, 225]

    plt.bar(products, sales, color=['b', 'g', 'r', 'c', 'y'])

    plt.title('Sales Data')
    plt.ylabel('Sales')

    plt.show()
bar_chart()
```

```python
def stacked_bars():
    times = ['T1', 'T2', 'T3', 'T4']
    plt.figure(figsize=(12, 8))
    category_A = np.array([130, 120, 65, 110])
    category_B = np.array([30, 10, 12, 60])
    category_C = np.array([20, 30, 25, 5])

    plt.bar(times, category_A, label='Category A', color='blue')
    plt.bar(times, category_B, bottom=category_A, label='Category B', color='green')
    plt.bar(times, category_C, bottom=category_A + category_B, label='Category C', color='red')

    plt.title('By Categories')
    plt.xlabel('Periods')
    plt.ylabel('Values')

    plt.legend()

    plt.show()
stacked_bars()
```
